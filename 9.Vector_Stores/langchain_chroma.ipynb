```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
# Create LangChain documents for Modern AI personalities
doc1 = Document(
    page_content="Geoffrey Hinton, often called the 'Godfather of Deep Learning', pioneered neural networks and backpropagation algorithms. His work laid the foundation for modern deep learning techniques used in vision and language models.",
    metadata={"field": "Deep Learning"}
)

doc2 = Document(
    page_content="Yann LeCun, a Turing Award winner, is known for his work on Convolutional Neural Networks (CNNs). He has contributed significantly to computer vision and is the Chief AI Scientist at Meta.",
    metadata={"field": "Computer Vision"}
)

doc3 = Document(
    page_content="Andrew Ng is a leading figure in AI education and applied machine learning. He co-founded Google Brain, led Baidu's AI Group, and has made AI accessible globally through online courses.",
    metadata={"field": "AI Education & Applied ML"}
)

doc4 = Document(
    page_content="Fei-Fei Li is a renowned computer scientist specializing in computer vision and cognitive neuroscience. She spearheaded the ImageNet project, which revolutionized large-scale visual recognition.",
    metadata={"field": "Computer Vision & Cognitive AI"}
)

doc5 = Document(
    page_content="Demis Hassabis is the co-founder and CEO of DeepMind, known for developing AlphaGo and other groundbreaking AI systems that combine deep learning with reinforcement learning.",
    metadata={"field": "Reinforcement Learning & AI Research"}
)
```

```python
docs = [doc1, doc2, doc3,   doc4, doc5]
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

vector_store = Chroma(
    embedding_function = embeddings,
    persist_directory='my_chroma_db',
    collection_name='sample'
)
```

```python
# add documents
vector_store.add_documents(docs)
```

```python
# view documents
vector_store.get(include=['embeddings','documents', 'metadatas'])
```

```python
# search documents
vector_store.similarity_search(
    query='Who among these are a computer vision expert?',
    k=2
)
```

```python
# search with similarity score
vector_store.similarity_search_with_score(
    query='Who among these are a computer vision expert?',
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query="",
    filter={"filed": "Computer Vision"}
)
```

```python
# update documents
updated_doc1 = Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore (RCB), is renowned for his aggressive leadership and consistent batting performances. He holds the record for the most runs in IPL history, including multiple centuries in a single season. Despite RCB not winning an IPL title under his captaincy, Kohli's passion and fitness set a benchmark for the league. His ability to chase targets and anchor innings has made him one of the most dependable players in T20 cricket.",
    metadata={"team": "Royal Challengers Bangalore"}
)

vector_store.update_document(document_id='paster_here', document=updated_doc1)
```
